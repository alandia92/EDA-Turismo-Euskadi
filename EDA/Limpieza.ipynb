```python
# Calcular la variación porcentual entre la primera y la última fila
var_perc = (df_12.iloc[-1, :] - df_12.iloc[0, :]) / df_12.iloc[0, :] * 100

# Ordenar los sectores de manera ascendente según su variación porcentual
var_perc_sorted = var_perc.sort_values(ascending=True)

# Mostrar los resultados
print(var_perc_sorted)
```

Para determinar los sectores más estables, es necesario calcular la variación porcentual o el cambio porcentual en el número de empleos en cada sector durante el período de tiempo considerado. Los sectores con una variación porcentual más baja o un cambio porcentual más estable se consideran los más estables.
Este código calcula la variación porcentual en cada sector entre enero de 2014 y febrero de 2023, y luego ordena los sectores de manera ascendente según su variación porcentual. Los sectores con una variación porcentual más baja o un cambio porcentual más estable se mostrarán primero.

Ten en cuenta que este enfoque asume que la evolución del empleo en cada sector ha sido constante y lineal a lo largo del tiempo, lo cual puede no ser el caso en la realidad. Por lo tanto, es posible que desees considerar otras medidas de estabilidad, como la variabilidad o la tasa de crecimiento promedio, en lugar de la variación porcentual en un período de tiempo específico.

```python
# Calcular la desviación estándar de los valores de empleo en cada sector
std_dev = df_12.std(axis=0)

# Ordenar los sectores de manera ascendente según su desviación estándar
std_dev_sorted = std_dev.sort_values(ascending=True)

# Mostrar los resultados
print(std_dev_sorted)
```

Para considerar otras medidas de estabilidad, como la variabilidad o la tasa de crecimiento promedio, puedes utilizar otras funciones estadísticas en lugar de la variación porcentual que calculamos en el ejemplo anterior.

Por ejemplo, la variabilidad se puede medir mediante la desviación estándar o la varianza de los valores de empleo en cada sector. Los sectores con una desviación estándar o una varianza más baja tendrán una evolución del empleo más estable.
Este código calcula la desviación estándar de los valores de empleo en cada sector y ordena los sectores de manera ascendente según su desviación estándar. Los sectores con una desviación estándar más baja tendrán una evolución del empleo más estable.

De manera similar, puedes calcular la tasa de crecimiento promedio para cada sector durante el período de tiempo considerado y ordenar los sectores según su tasa de crecimiento promedio. La tasa de crecimiento promedio se puede calcular como la tasa de crecimiento geométrico o como la tasa de crecimiento aritmético, dependiendo de tus necesidades y preferencias. Aquí te muestro un ejemplo de cómo puedes calcular la tasa de crecimiento aritmético y ordenar los sect

```python
import matplotlib.pyplot as plt
variance = df_12.var()

# Crear un gráfico de barras que muestra la varianza de los valores de empleo en cada sector
plt.bar(variance.index, variance.values)
plt.xlabel('Sector')
plt.ylabel('Varianza')
plt.title('Varianza del empleo por sector')
plt.show()
```

```python
fuente_20 = 'https://ourworldindata.org/travel-carbon-footprint'

df_20 = pd.read_csv('./data/asi_ha_aumentado_la_concentracion_de_co2_en_la_atmosfera.csv', sep=';')
df_20 = df_20.dropna(how='all')
df_20 = df_20.drop(df_20.index[345:])
df_20['Año'] = pd.to_datetime(df_20['Año'], format='%Y')
df_20 = df_20.set_index('Año')
df_20
```

```python

# Convertir la columna "CO2" en un valor numérico
df_20['CO2'] = df_20['CO2'].str.replace(',', '.').astype(float)

# Graficar la concentración de CO2 a lo largo del tiempo
ax = df_20.plot(figsize=(10, 6), legend=None)

# Resaltar las subidas y bajadas
for i in range(len(df_20)-1):
    if df_20.iloc[i+1]['CO2'] > df_20.iloc[i]['CO2']:
        ax.axvspan(df_20.index[i], df_20.index[i+1], color='red', alpha=0.2)
    elif df_20.iloc[i+1]['CO2'] < df_20.iloc[i]['CO2']:
        ax.axvspan(df_20.index[i], df_20.index[i+1], color='green', alpha=0.2)

# Configurar el título y los ejes
ax.set_title('Concentración de CO2')
ax.set_xlabel('Fecha')
ax.set_ylabel('CO2 (ppm)')

# Mostrar la gráfica
plt.show()


```

```python
df_20['Var %'] = df_20['CO2'].pct_change() * 100

# Marcar los períodos de subida y bajada
df_20['Subida'] = df_20['Var %'] > 0
df_20['Bajada'] = df_20['Var %'] < 0

# Mostrar los períodos de subida y bajada
print('Períodos de subida:')
print(df_20[df_20['Subida']])

print('\nPeríodos de bajada:')
print(df_20[df_20['Bajada']])
```
