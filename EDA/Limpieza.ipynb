Concentracion de CO2 en la atmosfera

```python
fuente_21 = 'https://ourworldindata.org/travel-carbon-footprint'

df_21 = pd.read_csv('./data/asi_ha_aumentado_la_concentracion_de_co2_en_la_atmosfera.csv', sep=';')
df_21 = df_21.dropna(how='all')
df_21 = df_21.drop(df_21.index[345:])

df_21
```

```python
def convertir_mes_a_numero(mes):
    fecha = datetime.datetime.strptime(mes, '%B')
    return fecha.month
# Aplicamos la función a la columna Periodo del DataFrame
df_21['numero_mes'] = df_21['Periodo'].apply(convertir_mes_a_numero)
```

```python
df_21['fecha'] = pd.to_datetime(df_21['Año'].astype(str) + df_21['numero_mes'].astype(str), format='%Y%m')
df_21.set_index('fecha', inplace=True)
```

```python
df_21 = df_21.drop(['Periodo', 'Año', 'numero_mes'], axis=1)
df_21
```
